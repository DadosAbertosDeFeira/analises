# Como está a alternância de poder entre os partidos no município?

Base de dados:
* Votação em partido por município e zona (em [Resultados no Repositório de dados eleitorais do TSE a partir de 1996](https://www.tse.jus.br/eleicoes/estatisticas/repositorio-de-dados-eleitorais-1/repositorio-de-dados-eleitorais))
* [Base dos Dados - Eleições](https://basedosdados.org/dataset/91292202-7188-4df1-a74d-e899e8fa824b)

Dados filtrados por Feira de Santana [aqui](https://docs.google.com/spreadsheets/d/1ndWim7eo7ws2LMECCVIuJNHQ1UuXL1Z6bqBbAE_EZpM/edit?usp=sharing).


```python
import pandas as pd


df = pd.read_csv("resultados_partido_municipio_feira_de_santana.csv")
df.head()
```

```python
df_first_round = df[df["turno"] == 1]
df_first_round
```

```python
df_transposed = df_first_round[["ano", "sigla_partido", "votos_nominais"]].pivot(index='ano', columns='sigla_partido', values='votos_nominais')
df_transposed.fillna(0)
```

```python
df_transposed.to_csv("votos_por_partido_por_ano.csv")
```

```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches


ax = df_transposed.plot.barh(stacked=True)

ax.set_title(
    "Quais os partidos disputaram as eleições para o poder executivo em Feira de Santana nos últimos 20 anos?",
    color='black'
)
ax.legend(bbox_to_anchor=(1.0, 1.0))
ax.plot()
```
