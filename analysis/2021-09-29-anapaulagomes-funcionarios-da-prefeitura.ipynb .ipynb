```python
import pandas as pd
```

```python
df = pd.read_csv("funcionarios.csv")
df.head()
```

```python
df["Tipo Servidor"].value_counts()
```

```python
import unicodedata


def strip_accents(string):
    if string is None:
        return
    return "".join(
        char
        for char in unicodedata.normalize("NFD", string)
        if unicodedata.category(char) != "Mn"
    )


positions = {}

print(len(df["Cargo"].value_counts()))

for key, value in df["Cargo"].value_counts().items():
    cleaned_position = strip_accents(key.upper())
    cleaned_position = cleaned_position.replace(" (A)", "")
    cleaned_position = cleaned_position.strip()
    if positions.get(cleaned_position):
        positions[cleaned_position] = positions[cleaned_position] + value
    else:
        positions[cleaned_position] = value

        
print("Valores antes da limpeza:", len(df["Cargo"].value_counts()))
print("Valores depois da limpeza:", len(positions))
```

```python
print("Cargo,Número de funcionários")

for key, value in positions.items():
    print(f"{key},{value}")
```
